```python
import random
from words import word_list

def get_word(): #returns a random word for the game
    word = random.choice(word_list)
    return word.upper() 
get_word()


def game_on(word):
        guessed_letters = []
        wrong = []
        tries = 6
        print("Let's play!")
        print(hangman_stages(tries))
        print("\n")

        while tries > 0:
            blank = ""
            for letter in word:
                if letter in guessed_letters:
                    blank = blank + letter
                else:
                    blank = blank + "-"

            if blank == word:
                break

            print("Guess the word:  ", blank)
            print(tries, "chances left.")
            print("\n")

            guess = input("Your guess:").upper()

            if len(guess) == 1 and guess.isalpha():
                if guess in guessed_letters or guess in wrong:
                    print(guess, "already guessed, try again!")
                elif guess in word:
                    print("Good job!", guess, "is in the word!")
                    guessed_letters.append(guess)
                else:
                    print("Sorry,", guess, "is not in the word.")
                    tries -= 1
                    wrong.append(guess)

                print(hangman_stages(tries))

            else:
                print("Not a valid guess.") 
                print("\n")
        if tries:
            print("YAY! You guessed the word! You win!!!")
        else:
            print("Sorry, you ran out of tries. The word was " + word + ".")
```

```python
def hangman_stages(tries): #the index of each stage corresponds to the number of tries the user has left
    HANGMAN_PICS = [  
                """
                   --------               
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
               
                """
                   --------
                   |      |              
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
               
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
               
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return HANGMAN_PICS[tries] 
```

```python
def main():
    word = get_word() 
    game_on(word) 
    while input("Play Again? (Y/N) ").upper() == "Y": #gives the user the option to play again
        word = get_word()
        game_on(word)
main()
```
